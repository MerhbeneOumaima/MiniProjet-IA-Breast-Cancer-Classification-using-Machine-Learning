Importing the Dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
data_frame = data_frame.dropna()
```

```python
# checking for missing values
print(data_frame.isnull().sum())
```

```python
data_frame = data_frame.drop_duplicates()
```

```python
# Vérification des doublons
print("Nombre de doublons :", data_frame.duplicated().sum())
```

```python
# Tracer des histogrammes pour chaque colonne du dataset
data_frame.hist(bins=20, figsize=(20, 15))
plt.tight_layout()  # Pour améliorer la disposition des graphiques
plt.show()
```

```python
# Calcul de la matrice de corrélation
correlation_matrix = data_frame.corr()

# Affichage de la matrice de corrélation avec une carte thermique (heatmap)
plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Matrice de Corrélation")
plt.show()
```

```python
# trouver les valeurs abarentes
for column in data_frame.columns[:-1]:
    sns.boxplot(data_frame[column])
    plt.title(f"Boxplot for {column}")
    plt.show()
```

```python
# statistical measures about the data
data_frame.describe()
```

Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
# training the Logistic Regression model using Training data

model.fit(X_train, Y_train)
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy on training data = ', training_data_accuracy)
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on test data = ', test_data_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Créer le modèle Random Forest
rf_model = RandomForestClassifier(random_state=42)

# Entraîner le modèle sur les données d'entraînement
rf_model.fit(X_train, Y_train)

# Faire des prédictions sur les données de test
rf_predictions = rf_model.predict(X_test)

# Calculer l'exactitude du modèle Random Forest
rf_accuracy = accuracy_score(Y_test, rf_predictions)
print("Accuracy of Random Forest Classifier:", rf_accuracy)
```

```python
from sklearn.svm import SVC

# Créer le modèle SVM
svm_model = SVC(kernel='linear', random_state=42)

# Entraîner le modèle sur les données d'entraînement
svm_model.fit(X_train, Y_train)

# Faire des prédictions sur les données de test
svm_predictions = svm_model.predict(X_test)

# Calculer l'exactitude du modèle SVM
svm_accuracy = accuracy_score(Y_test, svm_predictions)
print("Accuracy of SVM Classifier:", svm_accuracy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.svm import SVC

# Générer des données simulées
X, y = make_moons(n_samples=200, noise=0.2, random_state=42)

# Ajuster un modèle SVM non linéaire
model = SVC(kernel='rbf', C=1, gamma=1)
model.fit(X, y)

# Définir une grille pour la frontière de décision
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))

# Prédire les classes pour chaque point de la grille
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Tracer le graphique
plt.figure(figsize=(8, 6))

# Contour de la frontière de décision
plt.contour(xx, yy, Z, colors='green', levels=[0.5], linewidths=2, linestyles='-')

# Tracer les points
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Tumeur bénigne', edgecolor='k')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Tumeur maligne', edgecolor='k')

# Personnaliser le graphique
plt.title("Modèle de Classification", fontsize=14)
plt.xlabel("Volume", fontsize=12)
plt.ylabel("Densité", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
#amélioration des hyperparamètre
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
import numpy as np

# Définir un espace d'hyperparamètres à explorer
param_dist = {
    'C': np.logspace(-2, 2, 5),
    'kernel': ['linear', 'rbf'],
    'gamma': np.logspace(-3, 0, 4)
}

# Créer un RandomizedSearchCV
random_search = RandomizedSearchCV(SVC(random_state=42), param_dist, n_iter=10, cv=5, scoring='accuracy', random_state=42, verbose=1)
random_search.fit(X_train, Y_train)

# Meilleurs paramètres et performance
print("Best Parameters:", random_search.best_params_)
print("Best Cross-Validation Accuracy:", random_search.best_score_)
```
